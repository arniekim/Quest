# **QUEST 01: 숫자야구 만들기**

---

### **숫자야구: 숨겨진 정답 숫자를 맞히는 게임**

숨겨진 정답 숫자를 맞히는 게임. 사용자가 제시한 숫자 중 정답 숫자와 숫자만 일치하는 경우에는 ball,

숫자와 위치까지 일치하는 경우에는 strike를 출력하는 프로그램을 만들어보세요.

<br>

### **조건:**

 - 프로그램이 시작할 때 서로 다른 5개의 정답 숫자를 초기화합니다.

 - 사용자에게 5자리 숫자를 입력 받습니다.

 - 정답 숫자와 입력 숫자를 비교해서 ball과 strike를 출력합니다.

 <br>

### **예시:**

  - 정답: 39276 // 입력: 47209
  - 출력: >>> 1 strike 2 ball

  <br>

---

### **풀이**

1. 정답 숫자 생성 방법
2. 사용자 입력 처리 방법
3. 정답/입력값 비교 로직
4. 결과 출력 방법

<br>

```python
# 1. 정답 숫자 생성 방법
```

```python
import random

answer = random.sample(range(10), 5)  # [3, 9, 2, 7, 6] 형태

print(answer)
```

<br>

```python
# 2. 사용자 입력 처리 방법
```

```python
user =input("5개의 숫자를 입력하세요: ")

guess = []
for i in range(5):
    guess.append(int(user[i]))

print(guess)
```

<br>

```python
# 3. 정답/입력값 비교 로직
```

```python
strike = 0
ball = 0

for i in range(5):
    if guess[i] == answer[i]: # 같은 위치에 같은 숫자 → Strike
        strike += 1
    elif guess[i] in answer: # 위치는 다른데 숫자가 정답에 존재 → Ball
        ball += 1
```

<br>

```python
 # 4. 결과 출력
```

```python
print("정답:",answer)
print("입력:",guess)
print("결과:",strike, "strike", ball, "ball")
```
